## **CREATING DELTA MANAGED TABLES **
### Loading Dataset into the Bronze Schema

```python
%sql
CREATE SCHEMA IF NOT EXISTS insurance_cat.bronze
```

```python
%sql
drop table if exists insurance_cat.bronze.agents;
drop table if exists insurance_cat.bronze.claims;
drop table if exists insurance_cat.bronze.policies;
drop table if exists insurance_cat.bronze.customers;
drop table if exists insurance_cat.bronze.coverages;
drop table if exists insurance_cat.bronze.risk_assessments;
drop table if exists insurance_cat.bronze.payments;
```

```python
from pyspark.sql.functions import current_timestamp, current_date
```

```python
from pyspark.sql.functions import current_timestamp, current_date

(df_agent
 .withColumn("ingestion_timestamp", current_timestamp())
 .withColumn("ingestion_date", current_date())
 .write
 .format("delta")
 .mode("append")
 .option("mergeSchema", "true")
 .saveAsTable("insurance_cat.bronze.agents", partitionBy=["ingestion_date"])
)
```

```python
(df_claims
 .withColumn("ingestion_timestamp", current_timestamp())
 .withColumn("ingestion_date", current_date())
 .write
 .format("delta")
 .option("mergeSchema", "true")
 .mode("append")
 .saveAsTable("insurance_cat.bronze.claims", partitionBy=["ingestion_date"])
)
```
